```python
import torch
import time
from my_nano_gpt.my_nano import GPT,GPTConfig,DataLoaderLite
```

# training the model using Float32

```python
model=GPT(GPTConfig())

device=torch.device("cuda" if torch.cuda.is_available() else "cpu")

model.to(device)

train_dataloader=DataLoaderLite(B=4,T=1024,data_path="/kaggle/input/tiny-shakespear/input.txt")

print(f"the device is: {device}")

optimizer=torch.optim.AdamW(model.parameters())

#lets use tensorflot 32 rather than torch.float32
#torch.set_float32_matmul_precision('high')

for i in range(50):
    t0=time.time()  #this is the initial time
    optimizer.zero_grad()
    x,y=train_dataloader.next_batch()
    x,y=x.to(device),y.to(device)

    logits,loss=model(x,y)
    loss.backward()
    t1=time.time()
    optimizer.step()

    torch.cuda.synchronize()
    tdif=(t1-t0)*1000
    
    #how many tokens per second are we training on
    tokens_per_second=train_dataloader.B*train_dataloader.T
    
    print(F"the iteration:{i},the loss: {loss}: time:{tdif} ms, tokens per sec: {tokens_per_second}")
     
```

# training the model using tensorfloat 32

```python
model=GPT(GPTConfig())

device=torch.device("cuda" if torch.cuda.is_available() else "cpu")

model.to(device)

train_dataloader=DataLoaderLite(B=4,T=1024,data_path="/kaggle/input/tiny-shakespear/input.txt")

print(f"the device is: {device}")

optimizer=torch.optim.AdamW(model.parameters())

#lets use tensorflot 32 rather than torch.float32
torch.set_float32_matmul_precision('high')

for i in range(50):
    t0=time.time()  #this is the initial time
    optimizer.zero_grad()
    x,y=train_dataloader.next_batch()
    x,y=x.to(device),y.to(device)

    logits,loss=model(x,y)
    loss.backward()
    t1=time.time()
    optimizer.step()

    torch.cuda.synchronize()
    tdif=(t1-t0)*1000
    
    #how many tokens per second are we training on
    tokens_per_second=train_dataloader.B*train_dataloader.T
    
    print(F"the iteration:{i},the loss: {loss}: time:{tdif} ms, tokens per sec: {tokens_per_second}")
     
```
